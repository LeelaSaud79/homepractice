```python
from sklearn.impute import SimpleImputer
```

```python
mean_imputer=SimpleImputer(strategy='mean')
```

```python
df['cigsPerDay']=mean_imputer.fit_transform(df[['cigsPerDay']])
```

```python
df['totChol']=mean_imputer.fit_transform(df[['totChol']])
```

```python
df['glucose']=mean_imputer.fit_transform(df[['glucose']])
df['BPMeds']=mean_imputer.fit_transform(df[['BPMeds']])
df['BMI']=mean_imputer.fit_transform(df[['BMI']])
df['heartRate']=mean_imputer.fit_transform(df[['heartRate']])
```

```python
bins=np.arange(0,60,10)
```

```python
df['AgeBucket']=pd.cut(df.age,bins=bins)
```
